```python
# Import libraries
import pandas as pd
import numpy as np

# Example DataFrames
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Score': [90, 85, 88, 95, 92]
})

df2 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Frank', 'Grace'],
    'City': ['NY', 'LA', 'Chicago', 'Houston']
})

df_with_nan = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, None],
    'Score': [90, None, 88]
})

df_duplicates = pd.DataFrame({
    'A': [1, 1, 2, 3, 3],
    'B': [3, 3, 4, 5, 5]
})

# Data Structures
# 1. Series
series = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print("\n1. Series:")
print(series)

# 2. DataFrame
print("\n2. DataFrame:")
print(df)

# Data Manipulation
print("\n3. Head (First 3 rows):")
print(df.head(3))

print("\n4. Tail (Last 2 rows):")
print(df.tail(2))

print("\n5. Info:")
print(df.info())

print("\n6. Describe:")
print(df.describe())

print("\n7. GroupBy (Mean of Score by Age):")
print(df.groupby('Age').mean(numeric_only=True))

print("\n8. Merge:")
merged_df = pd.merge(df, df2, on='Name', how='inner')
print(merged_df)

print("\n9. Concat:")
concatenated_df = pd.concat([df, df2], axis=0)
print(concatenated_df)

# Data Filtering
print("\n10. Loc (Specific row and column):")
print(df.loc[1, 'Name'])

print("\n11. Iloc (Specific row and column):")
print(df.iloc[1, 2])

print("\n12. Query (Age > 30):")
print(df.query('Age > 30'))

print("\n13. Boolean Indexing (Score > 90):")
print(df[df['Score'] > 90])

# Data Cleaning
print("\n14. Dropna:")
print(df_with_nan.dropna())

print("\n15. Fillna (Replace missing values):")
print(df_with_nan.fillna(0))

print("\n16. DropDuplicates:")
print(df_duplicates.drop_duplicates())

# Data Transformation
print("\n17. Rename Columns:")
renamed_df = df.rename(columns={'Name': 'Full Name'})
print(renamed_df)

print("\n18. Melt (Wide to Long):")
melted_df = pd.melt(df, id_vars=['Name'], value_vars=['Age', 'Score'])
print(melted_df)

print("\n19. Pivot (Long to Wide):")
pivoted_df = melted_df.pivot(index='Name', columns='variable', values='value')
print(pivoted_df)

# Data Analysis
print("\n20. Mean:")
print(df.mean(numeric_only=True))

print("\n21. Median:")
print(df.median(numeric_only=True))

print("\n22. Std (Standard Deviation):")
print(df.std(numeric_only=True))

print("\n23. Corr (Correlation Matrix):")
print(df.corr(numeric_only=True))
```
