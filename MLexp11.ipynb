```python
for filename in uploaded.keys():

    df = pd.read_excel(filename)


    df.replace(grade_mapping, inplace=True)


    updated_filename = 'Updated_' + filename
    df.to_excel(updated_filename, index=False)


    files.download(updated_filename)
    print(f"Updated file saved as: {updated_filename}")

print("All files processed and downloaded!")
```
