```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/diabetes.csv"
df = pd.read_csv(file_path)

# Set Seaborn style
sns.set_style("whitegrid")

# Create individual histograms for each numerical feature
features = df.columns  # Get all column names

plt.figure(figsize=(12, 10))  # Set figure size

# Loop through each feature and plot histogram
for i, feature in enumerate(features):
    plt.subplot(3, 3, i + 1)  # Create a 3x3 grid of subplots
    sns.histplot(df[feature], bins=5, kde=True, color='blue')  # Histogram with KDE
    plt.title(f"Distribution of {feature}")  # Title for each plot

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(25, 7))
sns.boxplot(data=df[['Glucose', 'BloodPressure', 'BMI', 'Age']], palette="Set2")
plt.title("Box Plot of Numerical Features", fontsize=16)
plt.show()

```

```python
# Pair Plot
sns.pairplot(df[['Glucose', 'BloodPressure', 'BMI', 'Age', 'Outcome']], hue='Outcome', palette='coolwarm')
plt.show()

```

```python
# Count Plot
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=df, hue='Outcome', palette="viridis", legend=False)
plt.title("Diabetes Outcome Distribution", fontsize=16)
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Count")
plt.show()

```

```python
# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.scatterplot(x=df["Glucose"], y=df["BloodPressure"], color="orange", alpha=0.7)
plt.title("Relationship between Glucose and Blood Pressure", fontsize=14)
plt.xlabel("Glucose Level")
plt.ylabel("Blood Pressure")
plt.show()
```
